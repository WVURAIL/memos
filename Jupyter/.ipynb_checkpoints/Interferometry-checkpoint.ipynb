# Convolution, Auto-correlation and Cross-correlation


## Convolution

"In mathematics (and, in particular, functional analysis) convolution is a mathematical operation on two functions (f and g) to produce a third function, that is typically viewed as a modified version of one of the original functions, giving the integral of the pointwise multiplication of the two functions as a function of the amount that one of the original functions is translated" from wikipedia

$$
(f * g)(x) \,=\, \int_{-\infty}^{+\infty} f(x-t)\,g(t)\,dt
$$

*for multiple dimensions*

$$
\begin{align}
(f * g)(x_1,\ldots,x_n ) \,&=\, (f * g)({\bf x})\\
\,&=\, \int_{-\infty}^{+\infty} \ldots \int_{-\infty}^{+\infty} f(x_1-t_1, \ldots , x_n-t_n)\,g(t_1, \ldots, t_n) \,d^nt\\
\,&=\, \int_{-\infty}^{+\infty} f({\bf x}-{\bf t})\,g({\bf t}) \,d^nt\end{align}
$$

#### Properties of convolution 

$$
\begin{align}
f * g \,&=\, g * f&\qquad (\text{commutativity})\\
(f * g)* h \,&=\, f * (g * h)&\qquad (\text{associativity})\\
f * (g + h) \,&=\, (f * g) + (f * h) &\qquad (\text{distributivity})\\
(a\, g) * h \,&=\, a \, (g * h)&\qquad (\text{associativity with scalar multiplication})\\
\end{align}
$$

```python
def boxcar(x,a,b,amp):
    return amp*np.logical_and(x>=a,x<=b)

a= -0.2
b = 0.2
# Plotting range
xrange = [-2., 2.]
xpoints = 1000
   
# Resolution element
dx = (xrange[1]-xrange[0])/float(xpoints)

x = np.linspace(xrange[0], xrange[1], xpoints)
y = boxcar(x, a, b, 1.)


amp = float(xpoints)/((xrange[1]-xrange[0])* y.sum())
y = boxcar(x, a, b, amp)

result0 = np.convolve(y,y,'same')
result0 = dx*result0
        
fig = plt.figure()
ax1  = fig.add_subplot(311)
plt.subplots_adjust(hspace = 0.5)
    
ax1.grid(True)

ax1.plot(x,y,'b-')

plt.title("Boxcar function", fontsize=14,color='black')
    
ax2  = fig.add_subplot(312, sharex=ax1) 


# Plot a grid
ax2.grid(True)

ax2.plot(x,result0,'b-')

plt.title("Boxcar  convolved with boxcar function", fontsize=14,color='black')
    
ax3  = fig.add_subplot(313, sharex=ax2) 


# Plot a grid
ax3.grid(True)

result1 = np.convolve(result0,y,'same')
result1 = dx*result1

ax3.plot(x,result1,'b-')

plt.title("Triangle wave convolved with boxcar function", fontsize=14,color='black')      
```

## Cross-correlation


$$
\begin{split}
(f\star g)(x) \,&=\, ({f_-}^*\circ g)(x)\\
&=\, \int_{-\infty}^{+\infty} f^*(t-x)\,g(t)\,dt\\
&\underset{t^\prime = t-x}{=}\, \int_{-\infty}^{+\infty} f^*(t^\prime)\,g(t^\prime+x)\,dt^\prime\\
\end{split}\qquad \text{,}
$$

$$
\begin{align}
(f\star g)(x_1,\ldots,x_n ) \,&=\, (f\star g)({\bf x})\\
&=\, ({f_-}^*\circ g)(x)\\
&=\, \int_{-\infty}^{+\infty} \ldots \int_{-\infty}^{+\infty} f^*(t_1-x_1, \ldots , t_n-x_n)\,g(t_1, \ldots, t_n) \,d^nt\\
\end{align}
$$


## Autocorrelation
$$
\begin{split}
R\{f\}(x) \,&=\, (f\star f)(x)\\
&=\, (f_-^* * f)(x)\\
&=\, \int_{-\infty}^{+\infty} f^*(t-x)\,f(t)\,dt\\
&\underset{t^\prime = t-x}{=}\, \int_{-\infty}^{+\infty} f^*(t^\prime)\,f(t^\prime+x)\,dt^\prime\\
\end{split}\qquad \text{.}
$$

### Why Interferometry

```python
def DishDiameter(wavelength=0.21):
    """Compute the diameter of an aperture as a function of angular resolution and observing wavelength"""
    angres=(15e-3/3600)
    c = 299792458. #spped of light, m/s
    freq = c/(wavelength)/1e6 #
    D = 1.22 * wavelength/np.radians(angres) # assuming a circular aperture

    print('\n')
    print('At a frequency of %.3f MHz (Lambda = %.3f m)'%(freq, wavelength))
    print('an aperture diameter of D = %f m'%D)
    print('is required to achieve an angular resolution of %f degrees / %f arcmin / %f arcsec'%(angres, angres*60, angres*3600))
    print('\n')
    
w = ipywidgets.interactive(DishDiameter, wavelength=(0.1e-6, 2, 1e-6))
display(w)
```

# 2 element interferometer

We can combine these signals:
* by addition to form a *sum* interferometer, or $\sum$-interferometer
* by multiplication to form a *product* interferometer, or $\prod$-interferometer.

## The $\sum$-interferometer

$$P ={2 V_0^2(1+ \cos\Delta \Phi)} $$

## The $\prod$-interferometer 

$$R_c= \langle V_1 V_2 \rangle_t$$ 
$$R_c =\frac{V_0^2}{2}\cos{\omega \tau}=\frac{V_0^2}{2}\cos \Delta \Phi$$ where $\tau$ is the time delay in the signal path


This is a cosine correlator, one can consider a sine correlator by introducing an artificial phase delay of $\frac{\pi}{2}$ in one of the two signal paths:

$$V_1=V_{01} \cos (\omega (t + \tau))  \quad V_2 = V_{02} \cos (\omega t + \frac{\pi}{2} )$$

\begin{eqnarray}
R&=&\langle V_{01} V_{02} \cos{(\omega t + \frac{\pi}{2})} \cos{[\omega (t + \tau) ]} \rangle_t\\
R&=& V_0^2 \frac{\langle \cos(2 \omega t + \tau + \frac{\pi}{2})+\cos (\omega \tau - \frac{\pi}{2})\rangle_t}{2}
\end{eqnarray}
$$R_s =\frac{V_0^2}{2}\sin{\omega \tau}$$


we can combine the two operations to obtain a complex correlator 

$$R = \int_{\Omega}  I_\nu(\mathbf{s}) e^{-\imath 2\pi \frac{\mathbf{b}\cdot\mathbf{s}}{\lambda}} d\Omega = V$$

```python
theta=np.linspace(0.,180.,1000)
wlambda=5

fringe=np.exp(-1j*2*np.pi*wlambda*np.cos(np.radians(theta)))

cosabs=np.abs(np.real(fringe))
sinabs=np.abs(np.imag(fringe))

plt.figure()
plt.axes(polar=True)
plt.xlabel('theta')
plt.plot(np.radians(theta),np.abs(cosabs),'b',label="Cosine correlator")
plt.plot(np.radians(theta),np.abs(sinabs),'r',label="Sine correlator")
plt.legend(loc="best")
```

# The U,V plane and sampling it

## Obtaining projected baselines

We define the axes of the $XYZ$ frame, the frame descibing the projected baselines at the observer posoition, as:

* the $X$-axis points towards $(H=0^\textrm{h}, \delta = 0^{\circ})$ ($\gamma$ the point where the vernal equinox crosses the local meridian)
* the $Y$-axis towards $(H=-6^\textrm{h}, \delta = 0^{\circ})$ (East)
* the $Z$-axis towards the North Celestial Pole (NCP).

\begin{equation}
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}
=
\begin{pmatrix}
\lvert \mathbf{b} \rvert \cos \delta \cos H\\
-\lvert \mathbf{b} \rvert \cos \delta \sin H\\
\lvert \mathbf{b} \rvert \sin \delta
\end{pmatrix}
= \lvert \mathbf{b} \rvert
\begin{pmatrix}
\cos L_a \sin \mathcal{E} - \sin L_a \cos \mathcal{E} \cos \mathcal{A}\nonumber\\ 
\cos E \sin \mathcal{A} \nonumber\\
\sin L_a \sin \mathcal{E} + \cos L_a \cos \mathcal{E} \cos \mathcal{A}\\
\end{pmatrix}
\end{equation}

Transforming the terrestrial XYZ frame to the celestial *uvw* reference frame. Let ($H_0$, $\delta_0$) be the point on the celestial sphere in the direction of of a source $\mathbf{s_0}$. We now define:

* the $u$-axis lies in the celestial equatorial plane, and points toward the hour angle $H_0-6^\text{h}$.
* the $v$-axis lies in the plane of the great circle with hour angle $H_0$, and points toward the declination $\frac{\pi}{2}-\delta_0$.
* the $w$-axis points in the direction of $\mathbf{s_0}$, the source.

The ($u$,$v$) axes form a 2-D plane perpendicular to $\mathbf{s_0}$.

\begin{equation}
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0\\ 
-\sin \delta_0 \cos H_0 & \sin\delta_0\sin H_0 & \cos\delta_0\\
\cos \delta_0 \cos H_0 & -\cos\delta_0\sin H_0 & \sin\delta_0\\
\end{pmatrix} 
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}
\end{equation}

\begin{equation}
\mathbf{b_\lambda}
= 
\frac{\mathbf{b}}{\lambda}
= 
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\frac{1}{\lambda}
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0\\ 
-\sin \delta_0 \cos H_0 & \sin\delta_0\sin H_0 & \cos\delta_0\\
\cos \delta_0 \cos H_0 & -\cos\delta_0\sin H_0 & \sin\delta_0\\
\end{pmatrix} 
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}
\end{equation}

```python
def baseline_to_xyz(lengthbaseline, elevation, azimuth, latitude):
    x = np.cos(latitude)*np.sin(elevation) - np.sin(latitude)*np.cos(elevation)*np.cos(azimuth)
    y = np.cos(elevation)*np.sin(azimuth)
    z = np.sin(latitude)*np.sin(elevation) + np.cos(latitude)*np.cos(elevation)*np.cos(azimuth)
    
    xyz = np.array([(x,y,z)])
    
    return lengthbaseline * xyz.T



def xyz_to_baseline(ha, dec):
    a1 = np.sin(ha)
    a2 = np.cos(ha)
    a3 = 0.
    
    b1 = -1*np.sin(dec)*np.cos(ha)
    b2 = np.sin(dec)*np.sin(ha)
    b3 = np.cos(dec)
    
    c1 = np.cos(dec)*np.cos(ha)
    c2 = -1*np.cos(dec)*np.sin(ha)
    c3 = np.sin(dec)
    
    return np.array([(a1,a2,a3),(b1,b2,b3),(c1,c2,c3)])


## returns the length of individual baseline and angle baseline makes to source( azimuth )
def baseline_angles(antennaPosition,lamb):
    na = len(antennaPosition) #number of antennas
    nbl = na*(na-1)/2 #number of independent baselines
    length_angle = np.zeros((nbl, 2))
    k = 0
    for i in range(na):
        for j in range(i+1, na):
            length_angle[k,0] = lamb**(-1)*np.sqrt((antennaPosition[i,0]-antennaPosition[j,0])**2 + (antennaPosition[i,1]-antennaPosition[j,1])**2)
            length_angle[k,1] = np.arctan2((antennaPosition[i,1]-antennaPosition[j,1]) , (antennaPosition[i,0]-antennaPosition[j,0]))
            k = k +1
    return length_angle
```

```python
def track_uv(listha, lengthbaseline, elevation, azimuth, latitude, dec, ntimeslots):
    
    UVW = np.zeros((ntimeslots, 3), dtype=float)
    for i in range(ntimeslots):
        UVW[i, :] = np.dot(xyz_to_baseline(listha[i], dec),baseline_to_xyz(lengthbaseline, azimuth, elevation, latitude)).T    
    return UVW 

def track_uv_freq(ha, listfreq,lengthbaseline, elevation, azimuth, latitude, dec, nfreqs):
    c=3e8
    UVW = np.zeros((nfreqs, 3), dtype=float)
    listlamb=c/listfreq
    for i in range(nfreqs):
        UVW[i, :] = np.dot(xyz_to_baseline(ha[0], dec),baseline_to_xyz(lengthbaseline*listlamb[0]/listlamb[i], azimuth, elevation, latitude)).T    
    return UVW 
    
def plotuv(antennaPos,L,dec,h,Ntimes,lamb):
    plt.clf()
    fig=plt.figure()
    B = baseline_angles(antennaPos,lamb)
    na = len(antennaPos) #number of antennas 
    nbl = na*(na-1)/2 #number of baselines
    maxuv=0.
    for i in range (nbl):
        uv = track_uv(h,B[i, 0], 0., B[i, 1], L, dec, Ntimes)/1e3;
        if uv.max() > maxuv : maxuv=uv.max()
        plt.plot(uv[:,0], uv[:,1], 'b.',ms=5)
        plt.plot(-uv[:,0], -uv[:,1], 'r.',ms=5)
    plt.xlabel('u (klambda)')
    plt.ylabel('v (klambda)')
    plt.title('uv coverage')
    mb = maxuv*1.1
    plt.axes().set_aspect('equal')
    plt.xlim(-mb,mb)
    plt.ylim(-mb,mb)
    
def plotuv_freq(antennaPosition,L,dec,h,Nfreqs,lamb0,df):
    plt.clf()
    fig=plt.figure()
    c=3e8
    #print "lamb0="+str(lamb0)
    tabfreq=c/(lamb0)+np.arange(Nfreqs)*df

    B = baseline_angles(antennaPosition,lamb0)
    
    na = len(antennaPosition)
    nbl = na*(na-1)/2
    maxuv=0.
    for i in range (nbl):
        uv = track_uv_freq(h, tabfreq,B[i, 0], 0., B[i, 1], L, dec, Nfreqs)/1e3;
        if uv.max() > maxuv : maxuv=uv.max()
        plt.plot(uv[:,0], uv[:,1], 'b.',ms=4,alpha=0.5)
        plt.plot(-uv[:,0], -uv[:,1], 'r.',ms=4,alpha=0.5)
    plt.xlabel('u (klambda)')
    plt.ylabel('v (klambda)')
    plt.title('uv coverage')
    mb = maxuv*1.1 #5*np.sqrt((uv**2).sum(1)).max()
    uv.shape
    plt.axes().set_aspect('equal')
    plt.xlim(-mb,mb)
    plt.ylim(-mb,mb)
```
